## <b> The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.</b>

# <b> Explore and analyze the data to discover key findings pertaining to terrorist activities. </b>

```python
# Keeping the necessary features and excluding the others
data = data[["iyear", "imonth", "iday", "country", "country_txt", "region_txt", "city", "success", "attacktype1_txt", "nkill", "propvalue", "targtype1_txt", "latitude", "gname", "longitude", "targsubtype1_txt", "target1", "weaptype1_txt", "weapdetail"]]
```

```python
data.head()
```

```python
#Renaming the Features
data.rename(columns = {'iyear' : 'Year', "imonth" : "Month", "iday" : "Day", "country" : "CountryNo",
                     "country_txt" : "Country", "region_txt" : "Region", "city" : "City", "propvalue" : "PropertyValue",
                     "attacktype1_txt" : "AttackType", "targtype1_txt" : "TargetType", "gname" : "GroupName",
                     "targsubtype1_txt" : "TargetSubType", "target1" : "Target", "nkill" : "NoOfKills",
                     "weaptype1_txt" : "WeaponType", "weapdetail" : "WeaponDetail"}, inplace = True)
```

```python
data.head()
```

```python
#Checking the null value in the features
data.isnull().sum()
```

```python
#Dealing with the nan values 
data["City"].fillna("Unknown", inplace = True)
data["NoOfKills"].fillna(0, inplace = True)
data["PropertyValue"].fillna(0, inplace = True)
data["Target"].fillna("Unknown", inplace = True)
```

```python
data.isnull().sum()
```

```python
#Visualizing attacks per year with CountPlot chart
plt.figure(figsize=(24,12))
sns.countplot(x = "Year", data = data)
plt.title("Attacks per Year")
```

```python
#Visualizing the attacks per year with success and failures data by bar chart
plt.figure(figsize=(24,12))
sns.countplot(x = "Year", hue = "success", data = data)
plt.title("Success vs Failure")
```
